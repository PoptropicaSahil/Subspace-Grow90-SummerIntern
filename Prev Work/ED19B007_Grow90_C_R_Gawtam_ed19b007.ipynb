```python
import json
import pandas as pd
from pandas.io.json import json_normalize #package for flattening json in pandas df
import spacy
from spacy.matcher import phrasematcher
```

```python
data = pd.read_json("data.json")
data.head()
```

```python
nlp=spacy.blank("en")
```

```python
# Listing down all the users id
data["user_id"].unique()
```

```python
# msg of a given user id
user_msg = data[data["user_id"]=="2ddd42ce-245c-4c98-a563-a45435038a67"]
user_msg.head()
```

```python
msg = user_msg["message"]
print(user_msg.message)

msg = msg.to_string()
nlpmsg = nlp(msg)
```

```python
matcher = spacy.matcher.PhraseMatcher(nlp.vocab)

kwds = ["spent", "debited", "charged", "paid", "sent", "txn", "billed", "transaction", "credited"]

c = [nlp(item)for item in kwds]
matcher.add('keyword',c)
matches = matcher(nlpmsg)
```

```python
from collections import defaultdict

msg = defaultdict(list)
msg_id = defaultdict(list)
finalid = defaultdict(list)
finalmsg = defaultdict(list)

for idx, review in user_msg.iterrows():
    doc = nlp(review.message)
    matches = matcher(doc)
    founditems = set([doc[match[1]:match[2]] for match in matches])

    for item in founditems:
        msg[item].append(review.message)
        msg_id[item].append(review.user_message_id)
        finalid[str(item)].append(msg_id[item])
        finalmsg[str(item)].append(msg[item])

    
# messages having the specified keyword
# print(finalmsg["credited"])
# ids having the spec
print(finalid["credited"])
#print(msg_id)
```

```python
#for debited texts

import re

tl = finalmsg["debited"]
gt = []

for i in range(0, len(tl)):
    a = str(tl[i])

    f = re.findall('(?<=Rs)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    g = re.findall('(?<=Rs )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    b = re.findall('(?<=Rs.)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    d = re.findall('(?<=Rs. )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    c = re.findall('(?<=INR )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    e = re.findall('(?<=INR)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)

    gt.append(d)
    gt.append(b)
    gt.append(c)
    gt.append(e)
    gt.append(f)
    gt.append(g)

list_ofdeb = [ele for ele in gt if ele != []]
#print(list_ofdeb)
```

```python
sum = 0
for i in range(len(list_ofdeb)):
    r = str(list_ofdeb[i][0]).replace(',','')
    sum += float(r)
print (sum)
```
