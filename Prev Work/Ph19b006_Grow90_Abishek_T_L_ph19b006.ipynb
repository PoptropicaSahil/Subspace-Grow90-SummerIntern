```python
import spacy
from spacy.matcher import phrasematcher
```

```python
data = pd.read_json("data.json")
data.head()
```

```python
#Using nlp vocabulary language english
nlp=spacy.blank("en")
```

```python
# Listing down all the users id
data["user_id"].unique()
```

```python
# msg of a given user id
user_msg = data[data["user_id"]=="2ddd42ce-245c-4c98-a563-a45435038a67"]
user_msg.head()
```

```python
text = user_msg["message"]
print(user_msg.message)

a = text.to_string()
d = nlp(a)
```

```python
#Finding matcher of spacy package and finding matches
matcher = spacy.matcher.PhraseMatcher(nlp.vocab)

keywords = ["spent", "debited", "charged", "paid", "sent", "txn", "billed", "transaction", "credited"]

c = [nlp(item)for item in keywords]
matcher.add('keyword',c)

matches = matcher(d)

print(d[matches[0][1]:matches[0][2]])
```

```python
#creating a defaultdict and creating a list using a keyword to grouping the messages with same keywords
from collections import defaultdict

mesg = defaultdict(list)
mesg_id = defaultdict(list)
#print(item_ratings)

final_id = defaultdict(list)
final_msg = defaultdict(list)

for idx, review in user_msg.iterrows():
    doc = nlp(review.message)
    matches = matcher(doc)


    # Create a set of the items found in the review text
    founditems = set([doc[match[1]:match[2]] for match in matches])

    #print(found_items)
    for item in founditems:
        mesg[item].append(review.message)
        mesg_id[item].append(review.user_message_id)
        final_id[str(item)].append(mesg_id[item])

        final_msg[str(item)].append(mesg[item])
        #print(str(item))
    
# the text has been grouped according to keywords. To see the messages and the msg id of a pparticular keyword, use final_msg (keyword) and final_id(keyword)
# An example using keyword 'paid' has been given below
print(final_msg["paid"])

#ids having the spec
print(final_id["paid"])

#print(msg_id)
```

```python
import re

# using regex we find symbol Rs, INR and find the amount adjacent which will guve us the currency denominations
deb = final_msg["debited"]
gt = []
print(len(deb))
for i in range(0, len(deb)):
    a = str(deb[i])

    # we have considered all the denominations and its variations of the currency
    b = re.findall('(?<=Rs)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    c = re.findall('(?<=Rs )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    d = re.findall('(?<=Rs.)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    e = re.findall('(?<=Rs. )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    f = re.findall('(?<=INR )\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    g = re.findall('(?<=INR)\d{1,10}(?:,\d{3})*(?:.\d+)?(?=\s)',a)

    gt.append(b)
    gt.append(c)
    gt.append(d)
    gt.append(e)
    gt.append(f)
    gt.append(g)

res = [ele for ele in gt if ele != []]
print(res)
print(len(res))
```

```python
#finding sum debitted
sum_deb = 0
for i in range(len(res)):
    r = str(res[i][0]).replace(',','')
    sum_deb += float(r)
print (sum_deb)
```

```python
#reading credited messages
cred = final_msg["credited"]
gt = []
print(len(cred))
for i in range(0, len(cred)):
    a = str(cred[i])

    # we have considered all the denominations and its variations of the currency
    b = re.findall('(?<=Rs.)\d{1,7}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    d = re.findall('(?<=Rs. )\d{1,7}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    c = re.findall('(?<=INR )\d{1,7}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    e = re.findall('(?<=INR)\d{1,7}(?:,\d{3})*(?:.\d+)?(?=\s)',a)
    f = re.findall('(?<=Rs)\d{1,7}(?:,\d{3})*(?:.\d+)?(?=\s)',a)

    gt.append(d)
    gt.append(b)
    gt.append(c)
    gt.append(e)
    gt.append(f)

res = [ele for ele in gt if ele != []]
print(res)

```

```python
#fniding sum credited
sum_cred = 0
for i in range(len(res)):
    r = str(res[i][0]).replace(',','')
    sum_cred += float(r)
print (sum_cred)
```

```python
#printing details
print (f"Total amount credited is {sum_cred} and total amount debited is {sum_deb}")
print (f"Total balance is {sum_cred-sum_deb}")
```
