# ***Regex pattern for filtering money***

Finding amount without INR and Rs prefixes 

using this for easy calculations

```python
def finds12_money(s):
    pattern=re.compile(r'(?:Rs\.?|INR)\s*(\d+(?:[.,]\d+)*)')
    return re.findall(pattern, s)
```

Finding amount with Rs and INR prefixes

```python
def finds1_money(s):
    pattern=re.compile(r'[rR][sS]\.?\s*[,\d]+\.?\d{0,2}|[iI][nN][rR]\.?\s*[,\d]+\.?\d{0,2}')
    return re.findall(pattern, s)
```

assumed that the amount of money which comes first in the message as credited or debited money 

as many messages contain remaining balance after credited or debited money

there are some NONE values as some messages contain key words but not any credited or debited money

```python

def making_list(a):#creating the list of obtained values after applying regex function 
    data=pd.DataFrame(a)
    kbu=pd.DataFrame(data['message'].values.tolist(),index=data.index)
    zl=kbu.values.tolist()
    return zl
def list_sli(z):#consedering only first value in the msg containing Rs or INR
    ml=[]
    for i in range(len(z)):
        ml.append(z[i][0:1])
    return ml
def single_lis(m):#converting the object type into string
    kl=[]
    for i in range(len(m)):
        kl.append(m[i][0])
    return kl
def removing_none(k):#Removing NONE values as some msgs contain keyword but not any information about credited or debited money
    bhl=[]
    for i in range(len(k)):
        if k[i]!=None:
            bhl.append(k[i])
    return bhl
def replace_commas(b):#removing comma's from type string containing only numbers as it becomes problem while converting string to float
    cake=[]
    for i in range(len(b)):
        cake.append(b[i].replace(',',''))
    return cake
def convert_int(n):#converting string values to float for additional calculations
    final=[]
    for i in range(len(n)):
        final.append(float(n[i]))
    return final
```

# **Function for Calculating debited money of a particular month**

```python
def money_deb_month(n):#n represents the integer number of the month in order starting with jan=1
    debited_money_data = debited_money[(debited_money['timestamp']).dt.month == int(n)]
    t=debited_money_data['message'].apply(lambda x: finds12_money(x))
    h=making_list(t)
    i=list_sli(h)
    j=single_lis(i)
    k=removing_none(j)
    l=replace_commas(k)
    m=convert_int(l)
    return sum(m)
```

# **Function for Calculating Credited money of a particular month**

```python
def money_cred_month(n):#n represents the integer number of the month in order starting with jan=1
    credited_money_data = credited_money[(credited_money['timestamp']).dt.month == int(n)]
    ab=credited_money_data['message'].apply(lambda x: finds12_money(x))
    h=making_list(ab)
    b=list_sli(h)
    c=single_lis(b)
    d=removing_none(c)
    e=replace_commas(d)
    f=convert_int(e)
    return sum(f)
```
